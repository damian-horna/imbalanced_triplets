```python
def config_to_dataframe(config):
    res_dict = {
        "dataset": [],
        "Units 1st layer": [],
        "Units 2nd layer": [],
        "Units 3rd layer": [],
        "Learning rate": [],
        "Batch size": [],
        "Epochs": [],
        "Gamma": []
    }
    
    for ds in ds_names:
        conf = config[ds]
        res_dict["dataset"].append(ds)
        res_dict["Units 1st layer"].append(conf['nn_config']['units_2nd_layer'])
        res_dict["Units 2nd layer"].append(conf['nn_config']['units_3rd_layer'])
        res_dict["Units 3rd layer"].append(conf['nn_config']['units_latent_layer'])
        res_dict["Learning rate"].append(conf['lr'])
        res_dict["Gamma"].append(conf['gamma'])
        res_dict["Batch size"].append(conf['batch_size'])
        res_dict["Epochs"].append(conf['epochs'])
        
    return pd.DataFrame.from_dict(res_dict)
```
